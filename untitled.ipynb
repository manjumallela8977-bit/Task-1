```python
def circle_area(radius):
    return 3.14159 * radius * radius

radius = 5
print("Area of the circle:", circle_area(radius))
```

```python
def largest(a, b, c):
    return max(a, b, c)

print("Largest number is:", largest(10, 25, 15))
```

```python
def string_length(text):
    count = 0
    for char in text:
        count += 1
    return count

# Example usage
input_string = "Hello World"
result = string_length(input_string)

print("String:", input_string)
print("Length of the string:", result)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Example usage
number = 7

if is_prime(number):
    print(number, "is a Prime number")
else:
    print(number, "is not a Prime number")
```

```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input"
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Example usage
n = 6
result = fibonacci(n)

print("Fibonacci number at position", n, "is:", result)
```

```python
def reverse_string(text):
    if text == "":
        return text
    else:
        return reverse_string(text[1:]) + text[0]

# Example usage
input_string = "Hello"
result = reverse_string(input_string)

print("Original String:", input_string)
print("Reversed String:", result)
```

```python
# Program to find sum from 1 to n

n = 5
sum = 0

for i in range(1, n + 1):
    sum += i

print("Value of n:", n)
print("Sum from 1 to", n, "is:", sum)
```

```python
# Program to count even numbers in a list

numbers = [1, 2, 3, 4, 5, 6, 8]
count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("List:", numbers)
print("Number of even elements:", count)
```

```python
# Program to print prime numbers from 1 to 100

for num in range(2, 101):
    is_prime = True

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```
